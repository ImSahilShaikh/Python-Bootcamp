```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional Method for Plotting

```python
plt.plot(x,y)
# These labels are just examples not actual ones
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.title("First Plot Example")
```

### Multi-plots on same canvas
subplot() method takes three arguments as: number of rows, number of columns, plot number

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

## Object Oriented Method to plot

- creating an object <br>
add_axes([left of axes, bottom of axes, width , height]

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X-label")
axes.set_ylabel("Y-label")
axes.set_title("This is Object Oriented Way")
```

#### Multi plot on same canvas using object oriented method

```python
fig = plt.figure()          #creates an empty canvas

#add axes to canvas

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("BIG DUDE")
axes2.plot(y,x)
axes2.set_title("LITTLE NIBBA")
```

#### Subplots using object oriented method

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
```

To fix the overlapping we can use following latout

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fix,axes = plt.subplots(nrows=1,ncols=2)
```

the axes in above statement is just an array of all the axes which are added by us by specifying nrows and ncols instead of using fig.add_axes() method, can be seen by:


```python
axes
```

So we can iterate through it and plot the graphs

fix,axes = plt.subplots(nrows=1,ncols=2)

for cur_ax in axes:
    cur_ax.plot(x,y)

We can also index it to plot like follows :

```python
fix,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

## Figure Size and DPI
To set the figure size we have to pass a parameter while creating a figure object, figsize and initialize it with a tuple which will set its width and height of figure(width,height) in inches<br>
To set the DPI we can pass dpi parameter and set it to the value we want to use

```python
fig = plt.figure(figsize=(2,1),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Saving a figure
fig.savefig() method is used to save the figure <br>
pass a string to this method to give file name.<br>
We can also specify the dpi of figure which we want to save
NOTE: pass the string with the extension

```python
fig.savefig("MyPicture.png",dpi=800)
```

## Legends
To add legend we have to put the labels while plotting the graph<br>
We can pass loc parameter to legend to specify the location of the legend , pass 0 for best location<br>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="X Squared")
ax.plot(x,x**3,label="X Cubed")

ax.legend(loc = 0)
```

## Plot Appearance 
### Color <br>
To change the color of the plot we can pass color parameter as a string to plot method as follows <br>
Also hex codes can be used for custom colors

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='green')
```

### Line width & Line style
- Width you can pass linewidth whole or just lw

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',linewidth=4)
```

- Tranparent Line

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',lw=4,alpha = 0.2)
```

- Line Style<br>
Pass linestyle parameter as a string, shorthand is ls

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',lw=4,linestyle='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',lw=4,linestyle='-.')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',lw=4,ls=':')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x,y,color='purple',lw=4,ls = "steps")
```

### Markers

Marking all the x points which we have <br>
Pass new argument marker with string value to set type of marker, markersize/ms to set size of marker

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,c='red',lw=1,marker='o',ms=10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,c='red',lw=1,marker='*',ms=10)
```

Marker face color, marker edge width, marker edge color are also parameters

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,c='red',lw=1,marker='*',ms=20, markerfacecolor='yellow', markeredgewidth=3,markeredgecolor='purple')
```

## Control over axis Apperance i.e Plot Range
Setting x and y limits to show a particular plot

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
